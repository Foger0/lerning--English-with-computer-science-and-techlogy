<div>Hi, in this pages, I want to distinguish two types of function.<br>     
First and foremost, allow me to introduce the concept of "pure functions".</div><details>
<summary>纯函数具有仅接受输入并产生输出的属性，除了返回值外没有任何影响。</summary>
Pure functions have the property of only taking input and producing output, without having any side effects beyond returning a value.
</details>
Now let me give you a simple but typical example first. "abs()" a built-in function, which computes the absolute value of its argument is a pure function.

```python
abs(-58)
```

<details>
<summary>如你所见，我输入了一个值“-58”，但它仅返回了一个输出结果“58”，没有其他任何内容。</summary>
As you can see, I provided an input value of "-58", but it only returned an output of "58" without anything else.
</details>
<details>
<summary>这是另一个有用的示例。内置纯函数 "zip" 接受两个或多个可迭代对象（如列表或元组）作为其参数。然后，它按顺序组合这些可迭代对象的元素，并返回一个 "zip" 类的对象，该对象是一个可以转换为列表或元组的迭代器。</summary>
Here is another useful example. The built-in pure function called "zip" can take two or more iterable objects(such as Lists or tuple) as it argument,  it then combines the element of these iterable objects in sequence, and returns an object name of "zip" class, which is an iterator that can be transformed into lists or tuples.
</details>


```python
names = ["Alice", "Bob", "Cathy"]
ages = [25, 30, 22]

zipped = zip(names, ages)
print(type(zipped))
for name, age in zipped:
    print(f"{name} is {age} years old.")
```

<details>
<summary>相对于纯函数，非纯函数在每次使用时可能产生一些副作用，从而改变计算机的状态。读写文件的功能是它的常见组成部分。</summary>
In contrast to pure functions, non-pure fountions can generate some side effects, causing change to state of computer. Founctions for wrting and reading files are common example of it.  
</details>
At the moment, I would like to offer an entertaining example to illsurate the concept of the non-pure function.


```python
five=print(5)
print(five)
```

In this example, you can obviously see "print()" just taking an output on the sceem and returning nothing.

<summary>在最后部分，你应该了解纯函数和非纯函数各自的优缺点。尽管纯函数有着严格的条件限制，不能产生副作用或随时间改变行为，但它们可以更可靠地组合成嵌套表达式和复合调用表达式。在上面的非纯函数示例中，我们可以看到 print 在操作数表达式中没有返回有用的结果。非纯函数通常在一些特殊方面被使用。此外，调试纯函数要容易得多。基于这些原因，我们最好更多地设计纯函数。</summary>

In the final part, you should uderstand advantages and disadvantages of both pure functions and non-functions. Although pure functions have strict conditions, such as not yield side effects or change behavior overtime, they can be composed more reliably into nested expressions and compound call expressions. As seen in the non-pure function example above that print does not return a useful result when used in an operand expression. The non-pure function always be used in some special situations. Moreover, it is much easier to debug the pure function. For these reasons, it is better to design more pure functions whenever possible.
